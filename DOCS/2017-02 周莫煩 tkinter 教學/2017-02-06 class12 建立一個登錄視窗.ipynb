```python
import tkinter as tk
from tkinter import messagebox
import pickle

window = tk.Tk()
window.geometry('450x300')
window.title('Welcome!!!')

#welcome image 插入歡迎圖片
canvas = tk.Canvas(window, width=500, height=200)
image_file = tk.PhotoImage(file='welcome.gif')   #放在同一資料夾，一定要gif檔
image = canvas.create_image(0, 0, anchor='nw', image=image_file)
canvas.pack(side='top')

#user infomation
tk.Label(window, text='User Name:').place(x=50, y=150)  #這時都還是定錨在'nw'
tk.Label(window, text='Password:').place(x=50, y=190)

#entry infomation
var_user_name = tk.StringVar()
var_user_name.set('example@test.com')
entry_user_name = tk.Entry(window, textvariable=var_user_name).place(x=160, y=150)

var_password =tk.StringVar()
entry_password = tk.Entry(window, textvariable=var_password, show='*').place(x=160, y=190)

#login and sing up button
def user_login():
    #先將輸入的資料，另存
    user_name = var_user_name.get()
    user_password = var_password.get()
    # 先嘗試讀取註冊資料，看檔案存不存在（是不是有人註冊過）
    try:
        with open('users_info.pickle', 'rb') as user_file:
            user_info = pickle.load(user_file)
    #如果註冊資料不存在，就創建一個，並存入一個admin（預設管理員）
    except:
        with open('users_info.pickle', 'wb') as user_file:
            user_info = {'admin':'admin'}
            pickle.dump(user_info, user_file)
    
    #如果有讀到檔了，開始比對
    if user_name in user_info:
        if user_password == user_info[user_name]:
            tk.messagebox.showinfo(title='Welcom', message='Hallo ! ' + user_name) #名字符合，密碼正確
        else:
            tk.messagebox.showerror(title='Error', message='Your password is wrong, try again!') #名字符合，密碼錯誤
    else:
        should_sign_up = tk.messagebox.askyesno(message='You have not sign up yet, sign in today?')# 名字不符合
        if should_sign_up:
            user_sign_up() #引導至註冊視窗

       
    
def user_sign_up():
    def sign_up_to_window():
        nn = var_new_name.get()
        np = var_new_pass.get()
        cp = var_confirm_pass.get()
        try:
            with open('users_info.pickle', 'rb') as user_file:
                exist_user_file = pickle.load(user_file)
        #下面except自己加上來的，因為如果沒有按過login的話，不會有users_info.pickle，如果這邊只有讀檔會因為沒有檔案而報錯！
        except:
            with open('users_info.pickle', 'wb') as user_file:
                exist_user_file = {'admin':'admin'}               #因為沒有檔案的話也是直接創建並加入管理員（admin）
                pickle.dump(exist_user_file, user_file)
            
        #先判別輸入的np和cp是否不相同，不同的話要跳出錯誤視窗
        if np != cp:
            tk.messagebox.showerror(window_sign_up, message='Error! Password and confirm password must be same!')
        elif nn in exist_user_file:
            tk.messagebox.showerror(window_sign_up, message='User name already exist!')
        else:
            exist_user_file[nn] = np
            with open('users_info.pickle', 'wb') as user_file:
                pickle.dump(exist_user_file, user_file)
            tk.messagebox.showinfo(window_sign_up, message='Welcome! You have successfully signed up! ')
            window_sign_up.destroy()
                
    #創建另一個註冊視窗
    window_sign_up = tk.Toplevel(window) #新東西，另開視窗
    window_sign_up.title('sign up window')
    window_sign_up.geometry('350x200')
    
    #註冊視窗的結構
    tk.Label(window_sign_up, text='New name:').place(x=10, y=10)
    tk.Label(window_sign_up, text='New Password:').place(x=10, y=50)
    tk.Label(window_sign_up, text='Confirm Password:').place(x=10, y=90)
    
    var_new_name = tk.StringVar()
    var_new_name.set('example@test.com')
    new_name = tk.Entry(window_sign_up, textvariable=var_new_name)
    new_name.place(x=150, y=10)
    
    var_new_pass = tk.StringVar()
    new_pass = tk.Entry(window_sign_up, textvariable=var_new_pass, show='*')
    new_pass.place(x=150, y=50)
    
    var_confirm_pass = tk.StringVar()
    confirm_pass = tk.Entry(window_sign_up, textvariable=var_confirm_pass, show='*')
    confirm_pass.place(x=150, y=90)
    
    btn_confirm = tk.Button(window_sign_up, command=sign_up_to_window, text='Sign up') 
    btn_confirm.place(x=150, y=130)
    #按下註冊按鈕後，要在定義一個註冊資料的判別（sign_up_to_window）
    


btn_login = tk.Button(window, command=user_login, text='Login')
btn_login.place(x=170, y=260)

btn_sign_up = tk.Button(window, command=user_sign_up, text='Sign up')
btn_sign_up.place(x=270, y=260)

window.mainloop()
```
