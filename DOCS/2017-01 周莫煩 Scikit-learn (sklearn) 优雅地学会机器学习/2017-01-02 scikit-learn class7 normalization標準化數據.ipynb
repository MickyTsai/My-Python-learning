## 標準化數據 
就是將各種Ｘ欄位的資料，如因取樣範圍的數值各自差異過大，將其統合，便於機器學習  
在scikit-learn中是在preprocessing (預處理)

```python
import numpy as np
from sklearn import preprocessing

a = np.array([[10, 2.7, 3.6],
             [-100, 5, -2],
             [120, 20, 40]], dtype=np.float64)

print(a)                      #先看看為標準化前數據
print('---------------------------------------------')
print(preprocessing.scale(a)) #標準化過後
```

```python
from sklearn.cross_validation import train_test_split

#生成數據的其中一個模塊
from sklearn.datasets.samples_generator import make_classification

#scikit-learn的其中一種模型(新學)
from sklearn.svm import SVC
import matplotlib.pyplot as plt

X, y = make_classification(n_samples=300, n_features=2 , n_redundant=0, n_informative=2,
                           random_state=22, n_clusters_per_class=1, scale=100)

plt.scatter(X[:, 0], X[:, 1], c=y)
plt.show()
```

```python
#先試一次當數據沒有normalization時score準確率是多少

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
clf = SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```

```python
# 再試一次當有normalization時score準確率是多少

X = preprocessing.scale(X)  #normalization step
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.3)
clf = SVC()
clf.fit(X_train, y_train)
print(clf.score(X_test, y_test))
```
