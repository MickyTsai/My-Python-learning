```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation
%matplotlib nbagg

fig,ax = plt.subplots()  #class16 有上過，不過還是不是很了解

x = np.arange(0,2*np.pi,0.01)  #np.pi = 3.14
line, = ax.plot(x,np.sin(x))   #np.sin = sin

# 這部分看不太懂
def anima(i):
    line.set_ydata(np.sin(x+i/10))
    return line,
def init():
    line.set_ydata(np.sin(x))
    return line,


ani = animation.FuncAnimation(fig=fig,frames=100,interval=20,blit=True,func=anima,init_func=init)
#frames=幀數 
#interval=間隔，刷新毫秒數 
#blit=是否只更新有變化數值，False就會全部更新（mac要用False不然會bug)

plt.show()
```
