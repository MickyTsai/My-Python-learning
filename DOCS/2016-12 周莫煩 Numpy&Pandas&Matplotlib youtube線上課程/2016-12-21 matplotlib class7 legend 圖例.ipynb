```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(-3,3,50)
y1 = 2*x+1
y2 = x**2

plt.figure()


plt.xlabel('i am X')
plt.ylabel('i am Y')

plt.xlim((-1,2))
plt.ylim((-2,3))
x_newticks = np.linspace(-1,2,5)
plt.xticks(x_newticks)
plt.yticks([-2,-1,1,2,3],
           ['very bad','bad','normal','good','very good'])

#本節重點，所以將上方拉下來一起講
L1, = plt.plot(x,y1,label='up')  #要記得L1','有逗號，matplotlib中legend的特殊寫法
L2, = plt.plot(x,y2,linestyle='--',color='r',linewidth=3,label='down')
plt.legend(loc='best',handles=[L1,L2,],labels=['No1','No2']) #如果沒有用labels命名，會用plot中的名字。


plt.show()
```
