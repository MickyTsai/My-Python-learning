```python
import matplotlib.pyplot as plt
```

## 方法一 
`plt.subplot` 每個圖都要定義一次位置，分配位置的敘述較不清楚

```python
plt.figure()

#第一個圖，位置是[2,2]矩陣中的第一個
plt.subplot(2,2,1)
plt.plot([0,1],[0,1])
plt.title('Ax1')

#第二個圖，位置是[2,2]矩陣中的第二個
plt.subplot(2,2,2)
plt.plot([0,2],[0,2])
plt.title('Ax2')


#第三個圖，位置是[2,2]矩陣中的第三個
plt.subplot(223)                 #(223)這樣其實也能判讀位置
plt.plot([0,3],[0,3])
plt.title('Ax3')


#第四個圖，位置是[2,2]矩陣中的第四個
plt.subplot(224)
plt.plot([0,4],[0,4])
plt.title('Ax4')


plt.show()
```

## 方法二
`plt.subplot2grid` 每個圖都要定義一次位置，位置敘述較清楚  
rowspan：佔用多少行，預設1  
colspan：佔用多少列，預設1

```python

ax2_1 = plt.subplot2grid((3,3),(0,0),rowspan=1,colspan=3) #(3,3)矩陣中，從(0,0)位置開始
ax2_1.plot([0,1],[0,1])
ax2_1.set_title('AX2_1')

ax2_2 = plt.subplot2grid((3,3),(1,0),colspan= 2)
ax2_2.plot([0,2],[0,2])
ax2_2.set_title('Ax2_2')  #指令變成.set_title()，原本是plt.title()
ax2_2.set_xlim(-1,2)

ax2_3 = plt.subplot2grid((3,3),(1,2),rowspan= 2) 
ax2_3.plot([0,2],[0,2])
ax2_3.set_title('Ax2_3')

ax2_4 = plt.subplot2grid((3,3),(2,0))
ax2_4.plot([0,2],[0,2])
ax2_4.set_title('Ax2_4')

ax2_5 = plt.subplot2grid((3,3),(2,1))
ax2_5.plot([0,2],[0,2])
ax2_5.set_title('Ax2_5')

plt.show()
```

## 方法三
`gridspec`  先切好整個圖，再分配位置(反而覺得這個敘述更麻煩)


```python
#要額外import
import matplotlib.gridspec as gridspec

gs = gridspec.GridSpec(3,3)  #先切好圖

ax3_1 = plt.subplot(gs[0,:])  #敘述位置
ax3_1.plot([0,1],[0,1])
ax3_1.set_title('AX3_1')

ax3_2 = plt.subplot(gs[1,:2]) 
ax3_2.plot([0,1],[0,1])
ax3_2.set_title('AX3_2')

ax3_3 = plt.subplot(gs[1:,2])
ax3_3.plot([0,1],[0,1])
ax3_3.set_title('AX3_3')

ax3_4 = plt.subplot(gs[2,0])
ax3_4.plot([0,1],[0,1])
ax3_4.set_title('AX3_4')

ax3_5 = plt.subplot(gs[2,1])
ax3_5.plot([0,1],[0,1])
ax3_5.set_title('AX3_5')

plt.show()
```

## 方法四
`plt.subplots` 注意多一個“s” 把所有圖一次分配好位置，較快，定義方式比較特別

```python
#注意定義方式，後面就簡單
#f似乎是代表 f=plt.figure 
#((ax11,ax12),(ax21,ax22)) 似乎是(2,2)矩陣中 四個圖一次命名好
f,((ax11,ax12),(ax21,ax22)) = plt.subplots(2,2,sharex=True,sharey=True) 

ax11.scatter([2,1],[1,1])
ax11.set_title('Ax4_1')

ax12.scatter([1,1],[3,2])
ax12.set_title('Ax4_2')

ax21.scatter([1,2],[1,4])
ax21.set_title('Ax4_3')

ax22.scatter([0,1],[0,1])
ax22.set_title('Ax4_4')

plt.show()
```
