```python
import matplotlib.pyplot as plt
import numpy as np

n = 1024
x = np.random.normal(0,1,n)
y = np.random.normal(0,1,n)

T = np.arctan2(y,x) #for color value 單純問了圖表好看，先不用深究

plt.scatter(x,y,alpha=0.5,s=40,c=T)
plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)

plt.xticks(())  #隱藏x軸標籤
plt.yticks(())  #隱藏y軸標籤

plt.show()
```
