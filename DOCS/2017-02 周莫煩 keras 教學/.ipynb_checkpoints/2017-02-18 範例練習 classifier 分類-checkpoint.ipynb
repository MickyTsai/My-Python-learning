```python
import numpy as np

from keras.models import Sequential
from keras.layers import Dense, Activation
from keras.datasets import mnist #讀取範例資料
from keras.utils import np_utils 
from keras.optimizers import RMSprop


########## load data 讀取資料 ########## 
(X_train, y_train), (X_test, y_test) = mnist.load_data()
#print(X_train.shape[0])  60000
#print(X_train.shape[1])  28
#print(X_train.shape[2])  28
#print(y_train.shape[0])  60000
#print(X_test.shape[0])   10000
#print(X_test.shape[1])   28
#print(X_test.shape[2])   28
#print(y_test.shape[0])   10000 
#X_train[0]   是一個28x28二維矩陣，裡面的值0~255(代表深淺度？) 



########## data pre-processioning 數據預處理 ###########
X_train = X_train.reshape(X_train.shape[0], -1)/255  #先將三維轉為二維(60000, 784)，然後使每個值都處於(0~255)/255
X_test = X_test.reshape(X_test.shape[0], -1)/255

y_train = np_utils.to_categorical(y_train, nb_classes=10)  #將[0] [1] [2] 標籤轉為 [1, 0, 0] [0, 1, 0] [0, 0, 1]
y_test = np_utils.to_categorical(y_test, nb_classes=10)

####### another way to build neural net ######
model = Sequential([Dense(input_dim=784, output_dim=32), 
                    Activation('relu'), 
                    Dense(10), 
                    Activation('softmax')])

###### another to define optimizer #########
rmsprop = RMSprop(lr=0.001, rho=0.9, epsilon=1e-08, decay=0.0) 

##### get metrics to get more results you want to see #####
model.compile(optimizer=rmsprop, 
              loss='categorical_crossentropy', 
              metrics=['accuracy'])

###### another way to train model ######
print('training........')
model.fit(X_train, y_train, nb_epoch=2, batch_size=32)

##### evaluate the model with mertrics ######
print('testing.........')
loss, accuracy = model.evaluate(X_test, y_test)

print('test_loss:', loss)
print('test_accuracy:', accuracy)
```
