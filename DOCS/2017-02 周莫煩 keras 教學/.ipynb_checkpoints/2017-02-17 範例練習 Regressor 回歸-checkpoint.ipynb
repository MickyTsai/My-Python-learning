```python
import keras
import numpy as np
from keras.models import Sequential  #一種“按照順序”建立神經層
from keras.layers import Dense      #全連接的神經層結構
import matplotlib.pyplot as plt


########### create some data:建立數據 ############
X = np.linspace(-1, 1, 200)
#shuffle(randomize the data):洗牌（將數據隨機化）
np.random.shuffle(X)  
Y = 0.5*X + 2 + np.random.normal(0, 0.05, X.shape) #添加一些雜亂的數，較符合一般情況

################# plt:先畫出數據看看 ###############
plt.scatter(X, Y)
plt.show()

############ 200個數據中，前160當作訓練數據，後40個當作測試數據##########
X_train, Y_train = X[:160],Y[:160]
X_test, Y_test = X[160:], Y[160:]


######## build a neural network from 1st layer to the last layer: 建立神經層結構 ########
model = Sequential()
model.add(Dense(input_dim=1, output_dim=1))   #dim:dimension(尺寸)


###########choose loss function and optimzing method: 選擇誤差的方程和優化器 ###########
model.compile(loss='mse', optimizer='sgd')   #sgd=Stochastic gradient descent:隨機梯度下降


######### training:訓練 ###########
print('training........')
for step in range(301):
    cost = model.train_on_batch(X_train, Y_train)   #會有默認返回值
    if step % 100 == 0:
        print('train cost:', cost)
        
####### test:測試 #####
print('testing.....')
cost = model.evaluate(X_test, Y_test, batch_size=40)
print('test cost:', cost)
#也抓出訓練後的W和b，看是否和產生數據的方程參數接近（越相近代表訓練越好）
W, b = model.layers[0].get_weights()
print('Weight:', W, '\nbiases', b)


###### 畫出最後回歸出來的線 ######
Y_pred = model.predict(X_test)
plt.scatter(X_test, Y_test)
plt.plot(X_test, Y_pred, 'r')
plt.show()

```
