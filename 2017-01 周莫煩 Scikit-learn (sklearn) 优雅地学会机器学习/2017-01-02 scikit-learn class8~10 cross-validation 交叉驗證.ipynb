## 交叉驗證-1
當使用cross-validation的 train_test_split時  每次的test_data（30%的部分）所帶入madel得出的score值(準確率)會跳動。  
* 這次使用cross_val_score 一次取多次test_data再求準確率平均值
* 當帶入不同的模型參數時，我們來可視化參數帶入時的趨勢，來達到最佳化參數，也避免over-fit

```python
from sklearn.datasets import load_iris
from sklearn.cross_validation import train_test_split
from sklearn.neighbors import KNeighborsClassifier

iris = load_iris()
X = iris.data
y = iris.target

train_X,test_X,train_y,test_y = train_test_split(X,y,random_state=4)
knn = KNeighborsClassifier(n_neighbors=5)  #考慮的鄰近數值量，太多會有over-fit的情形
knn.fit(train_X,train_y)

#先看一下在n_neighbors=5 時準確率如何
print(knn.score(test_X,test_y))
print('-----------------------------------------')




#（重點一）一次創建多組test_data去帶入後平均
from sklearn.cross_validation import cross_val_score

scores = cross_val_score(knn,X,y,cv=5,scoring='accuracy')  
#cv=5 五組數據， scoring='accuracy'評準確率的一種方法，以後會學到其他種，不同模型有建議不同對應的。

print(scores) # 各組的準確率
print(scores.mean()) #  平均值




#（重點二）運用迭代以及可視化，來找到最佳參數
import matplotlib.pyplot as plt

k_range = range(1,31)
k_scores = []

for k in k_range:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn,X,y,cv=10,scoring='accuracy')
    k_scores.append(scores.mean())
    
plt.plot(k_range,k_scores)
plt.xlabel('Value of k for KNN')
plt.ylabel('Cross Validated Accuracy')
plt.show()
```

* 可以看出當n_neightbors參數介於12~18之間，得到的scores.mean準確率平均值最高。  
* 補充：當model是Regression時，scoring建議’mean_squared_error’（平均方差值-對比誤差值）

## 交叉驗證-2
* 上面是用最後的scores來看參數的最佳化  
* 這次用learning_curve，模型的學習曲線來看，當在何種學習情況（給的不同的train_size
）可以得到較小對比誤差值

```python
from sklearn.datasets import load_digits
from sklearn.svm import SVC

digits =  load_digits()
X = digits.data
y = digits.target




#（重點）使用learning_curve
from sklearn.learning_curve import learning_curve

import matplotlib.pyplot as plt
import numpy as np

train_sizes, train_loss, test_loss= learning_curve(
        SVC(gamma=0.001), X, y, cv=10, scoring='mean_squared_error',
        train_sizes=[0.1, 0.25, 0.5, 0.75, 1])
train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(train_sizes, train_loss_mean, 'o-', color="r",
             label="Training")
plt.plot(train_sizes, test_loss_mean, 'o-', color="g",
             label="Cross-validation")

plt.xlabel("Training examples")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```

* （綠）可以看出test_data在一開始（給的train_data少）誤差會高，之後不斷學習後，誤差就降低  
* （紅）由於給的和對比的都是train_data可見一開始學習就得到很高的準度
* 所以可以了解 train_data是讓model去學習。而test_data是去測試模型準度。
* 模型係數（例如：SCV中的gamma，knn的n_neighbors都會影響學習曲線）

## 交叉驗證-3
其實就是交叉驗證-1中的另一種模型在不同參數的模型可視化（參數曲線？）

```python
from __future__ import print_function
from sklearn.learning_curve import  validation_curve
from sklearn.datasets import load_digits
from sklearn.svm import SVC
import matplotlib.pyplot as plt
import numpy as np

digits = load_digits()
X = digits.data
y = digits.target


param_range = np.logspace(-6, -2.3, 5)
train_loss, test_loss = validation_curve(
        SVC(), X, y, param_name='gamma', param_range=param_range, cv=10,
        scoring='mean_squared_error')

train_loss_mean = -np.mean(train_loss, axis=1)
test_loss_mean = -np.mean(test_loss, axis=1)

plt.plot(param_range, train_loss_mean, 'o-', color="r",
             label="Training")
plt.plot(param_range, test_loss_mean, 'o-', color="g",
             label="Cross-validation")

plt.xlabel("gamma")
plt.ylabel("Loss")
plt.legend(loc="best")
plt.show()
```
