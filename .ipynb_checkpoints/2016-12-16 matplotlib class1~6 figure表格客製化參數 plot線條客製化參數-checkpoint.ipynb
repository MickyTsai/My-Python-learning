```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-2,2,10)
y1 = 2*x + 1
y2 =  x**2 

#分別用plot.show()隔開，各自一個表格。
plt.plot(x,y1)
plt.show()
plt.plot(x,y2)
plt.show()
```

```python
#直接連續輸入，就在同一表格
plt.plot(x,y1)   
plt.plot(x,y2)
plt.show()
```

```python
#用figure指令，客製化自己的表格。

plt.figure() #figure指令的下方會是此figure的參數，不用另外多打plt.show
plt.plot(x,y1)
plt.figure(num=3,figsize=(5,5))  #所以可以在figure中輸入各自參數[num=表格名稱，figsize=表格大小]
plt.plot(x,y2)
plt.show()
```

```python
#同一個figure 客製化各個線條
plt.figure(figsize=(5,5))
plt.plot(x,y1,color='red',linewidth=5,linestyle='--')
plt.plot(x,y2,color='Green',linewidth=10)
plt.show()
```

```python
#  x,y表格顯示範圍參數
plt.figure(figsize=(3,3))
plt.plot(x,y1,color='red',linestyle='--')
plt.plot(x,y2,color='green')

plt.xlim(-2,2)
plt.ylim(-5,5)

plt.xlabel('i am X')
plt.ylabel('i am Y')
plt.show()
```

```python
# 取代掉x,y 數據
plt.figure(figsize=(5,5))
plt.plot(x,y1,color='red',linestyle='--')
plt.plot(x,y2,color='green')

new_ticks = np.linspace(-2,2,5)
print(new_ticks)
plt.xticks(new_ticks)

plt.yticks([-5,-4,-1,4,5],
          ['very bad','bad','normal','good','very good'])
plt.show()
```

```python
#改變字型，數學符號
plt.figure(figsize=(5,5))
plt.plot(x,y1,color='red',linestyle='--')
plt.plot(x,y2,color='green')

new_ticks = np.linspace(-2,2,5)
print(new_ticks)
plt.xticks(new_ticks)

plt.yticks([-5,-4,-1,4,5],
          [r'$very\ bad$',r'$bad$',r'$\alpha$',r'$good$',r'$very\ good$']) 
plt.show()
```

```python
#移動x,y軸
plt.figure()
plt.plot(x,y1,color='red',linestyle='--')
plt.plot(x,y2,color='green')

new_ticks = np.linspace(-2,2,5)
plt.xticks(new_ticks)

plt.yticks([-5,-4,-1,4,5],
          [r'$very\ bad$',r'$bad$',r'$\alpha$',r'$good$',r'$very\ good$']) 

# gca = 'get current axis'
ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['bottom'].set_position(('data',0)) #定位的參數還有 [outward,axis,....]
ax.spines['left'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')

plt.show()
```
