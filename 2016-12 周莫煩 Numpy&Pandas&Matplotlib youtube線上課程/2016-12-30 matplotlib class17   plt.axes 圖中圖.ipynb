```python
import matplotlib.pyplot as plt

x = [1,2,3,4,5,6,7]
y = [3,6,8,9,2,4,1]

fig = plt.figure()

left,bottom,width,height = 0.1,0.1,0.9,0.9  #left,bottom圖表起始位置，width,height圖表跟原尺寸比例
ax1 = fig.add_axes([left,bottom,width,height])
ax1.plot(x,y,'r')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_title('ax1')

left,bottom,width,height = 0.15,0.7,0.25,0.25
ax2 = fig.add_axes([left,bottom,width,height])
ax2.plot(y,x,'g')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax2.set_title('ax2')

plt.axes([0.7,0.15,0.25,0.25])
plt.plot(x,y,'b')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('ax3')

plt.show()
```
