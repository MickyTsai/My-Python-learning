```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D #3D圖表
```

```python

#可讓figure有互動功能 
#來源：http://stackoverflow.com/questions/25333732/matplotlib-animation-not-working-in-ipython-notebook-blank-plot
%matplotlib nbagg  

fig = plt.figure()
ax = Axes3D(fig)

x = np.arange(-4,4,0.25)
y = np.arange(-4,4,0.25)

x,y = np.meshgrid(x,y)   #這個還是不太懂

r = np.sqrt(x**2+ y**2)  #np.sqrt 求平方根
z = np.sin(r)

#xyz都有了 開始畫3D圖
ax.plot_surface(x,y,z,rstride=1,cstride=3,cmap=plt.get_cmap('bone'))  #rstride,cstride 曲面上網格的長寬大小
plt.contourf(x,y,z,zdir='z',offset=-2,cmap='rainbow')  #zdir將等高線圖投影在哪個面，offset投影面上哪個值上投影

ax.set_zlim(-2,2) #Axes中，設定xlim,ylim,zlim有點不一樣
plt.show()
```
