```python
import matplotlib.pyplot as plt
import numpy as np

def f(x,y):
    # the height function xy數值帶入後得到高度的公式
    return(1-x/2+x**5+y**3)*np.exp(-x**2-y**2)

n = 256
x = np.linspace(-3,3,n)
y = np.linspace(-3,3,n)

#老實講著部分看不懂 mesh(網格)，好像轉換成等高線的數值吧？？？
X,Y = np.meshgrid(x,y)

#繪製等高線，等高線標籤(預覽)
#C = plt.contour(X,Y,f(X,Y)) 
#plt.clabel(C)

#use plt.contourf to filling contours 等高線上色填滿
#X,Y and value for (X,Y) point
plt.contourf(X,Y,f(X,Y),8,alpha=0.8,cmap=plt.cm.hot)  #8 代表切出幾個等高線部分 0=2部分 8=10部分

#use plt.contiur to add contour lines 繪出等高線的線
C = plt.contour(X,Y,f(X,Y),8,linewidth=.5,colors='black')

#adding label
plt.clabel(C,inline=True,fontsize=10) #inline=True 標籤貼在線上 False線會穿過標籤

plt.xticks(())
plt.yticks(())
plt.show()
```
