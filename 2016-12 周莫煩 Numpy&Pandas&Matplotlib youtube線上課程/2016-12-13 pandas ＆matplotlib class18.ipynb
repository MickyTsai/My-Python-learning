```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt

#Series

data = pd.Series(np.random.randn(100))
print(data)
```

```python
data = data.cumsum()
data.plot()
plt.show()
```

```python
# DataFrame

df = pd.DataFrame(np.random.randn(100,4),columns=list('abcd'))
print(df.head(20)) #只印出前20個數據
```

```python
df.plot()
plt.show()
```

```python
df = df.cumsum() #累加數據
df.plot()
plt.show()
```

```python
# plot methods: data.plot.xxx
# 'bar' 'hist' 'show' 'box' 'kde' 'area' 'scatter' 'hexbin' 'pie'
# 介紹'scatter' 點狀圖

ax1 = df.plot.scatter(x='a',y='b',color='Blue',label='class-1')
df.plot.scatter(x='a',y='c',color='Red',label='class-2',ax=ax1) #將此scatter結果和前一個ax合併一起繪出
plt.show()
```
