```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3,50)
y = 2*x + 1
plt.figure(num=1,figsize=(8,5))
plt.plot(x,y,color='r',linestyle='--',linewidth=2)

ax = plt.gca()
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
ax.spines['left'].set_position(('data',0)) #將y軸（left），依照（data）中的值0定位
ax.spines['bottom'].set_position(('data',0))
ax.xaxis.set_ticks_position('bottom')  #將某一個脊柱設為x軸（top,bottom都有可能）
ax.yaxis.set_ticks_position('left')

#要在x1,y=2(1)+1=3的地方做標注點和x軸對應線。

x0 = 1
y0 = 2*x0+1
plt.scatter(x0,y0,s=50,color='b') #先標上數據上的點  #s=點的大小
plt.plot([x0,x0],[y0,0],'b',lw=3)  #兩個點的x座標[x0,x0],兩個點的y座標[y0,0]，不是第一個點的xy座標和第二個點的xy座標。

# method 1
# 製作標注註解和註解線。
plt.annotate(r'$2x+1=%s$'%y0,
             xy=(x0,y0),xycoords='data',
            xytext=(30,-30),textcoords='offset points',fontsize=16,
            arrowprops=dict(arrowstyle='->',connectionstyle='arc3,rad=.2'))

#method 2 
#製作文字方塊加入表格
plt.text(-3.7,3,r'$This\ is\ some\ text.\ \mu\ \sigma_i\ \alpha_t$',
        fontdict={'size':15,'color':'r'})


plt.show()
```
