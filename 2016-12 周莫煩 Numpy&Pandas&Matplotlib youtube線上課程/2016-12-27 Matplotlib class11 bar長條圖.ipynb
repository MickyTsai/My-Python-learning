```python
import matplotlib.pyplot as plt
import numpy as np

n = 12
x = np.arange(n)
y1 =(1 - x/float(n)) * (np.random.uniform(0.5,1,n))
y2 =(1 - x/float(n)) * (np.random.uniform(0.5,1,n))

plt.bar(x,y1,facecolor='#9999ff',edgecolor='white') #可以打入色碼，選擇較多較好看
plt.bar(x,-y2,facecolor='#ff9999',edgecolor='white')
plt.xlim(-0.5,n)
plt.xticks(())
plt.ylim(-1.5,1.5)
plt.yticks(())

# 在長條圖顯示數值（較難）

for X,Y in zip(x,y1):
    plt.text(X+0.4,Y+0.02,'%.2f'%Y,ha='center',va='bottom') 
    #數值ＸＹ座標平移， 
    #ha:horizontal alignment 水平對校準 va:垂直校準（就是對準長條圖中心'center'左右'right''left頂部'top'或底部'bottom'）
    #'%.2f'%Y 帶入Ｙ數值，保留小數兩位。
    
for X,Y in zip(x,y2):
    plt.text(X+0.4,-Y-0.02,'%.2f'%Y,ha='center',va='top')

print(x)
print(y1)
print(-y2)
plt.show()
```
